```python
#wrapping above all
def get_xkcd_image(idx=1000):
    url='https://xkcd.com/{}/'.format(idx)
    
    r=requests.get(url)
    if r.status_code==200:
        soup=BS(r.text)
        img_url='https:' + soup.find_all('div', id='comic')[0].img['src']
    return image_url
```

```python
for ix in range(1,1000,100):
    print ix, get_xkcd_image(ix)
```
